```python
# Function to rename multiple files
def main():
    i = 0
    path="/home/neno/WebProjects/The Odin Project Projects/portfolio-TOP/public/img/thumbs/highres/"
    for filename in os.listdir(path):
        # Get file name
        image_name = filename[0:-4]
        print(image_name)
        # Get file format
        image_format = filename[-4:]
        print(image_format)
        if image_format == "pynb": continue
        print(filename)
        
        print("path ",path)
        
        # Set new name
        my_dest = image_name + "-highres" + image_format
        print(my_dest)
        # Set source of original file
        print("path ",path)
        print("filename",filename)
        my_source = path + filename
        print(my_source)
        # Set source new renamed file
        my_dest = path + my_dest
        # Rename the original file
        os.rename(my_source, my_dest)
# Driver Code
if __name__ == '__main__':
    # Calling main() function
    main()
```
